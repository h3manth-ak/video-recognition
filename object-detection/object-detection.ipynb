```python
from frame_extraction_obj import VideoFrameExtractor
```

```python
video_path = 'C:/Users/USER/Desktop/main-project/dataset/program/test.mp4'
frame_extractor = VideoFrameExtractor(video_path)

frames = frame_extractor.get_frames()
```

```python
counter = 0
for frame in frames:
    results = model.predict(frame)
    counter += 1
    print(counter)
    # Image(filename='', height=600)
```

```python
# from PIL import Image
from xttym import VideoTimestamp

counter = 0
names=model.names
ts=[]
video_timestamp = VideoTimestamp(video_path)
for frame in frames:
    counter += 1
    print(counter)
    results=model.predict(frame)
    for r in results:
        for c in r.boxes.cls:
            if names[int(c)] == 'car':
               ts.append(video_timestamp.get_timestamp(counter))  # Use the instance to get the timestamp
               

video_timestamp.release()
```

```python
ts
```

```python
import numpy as np
ts_updated=ts
ts_updated=np.array(ts_updated)
ts_updated_filtered = ts_updated[abs(ts_updated - np.mean(ts_updated)) < 2 * np.std(ts_updated)]
```

```python
from moviepy.video.io.ffmpeg_tools import ffmpeg_extract_subclip

# Specify the input video file path
input_video_path = video_path

# Specify the output video file path
output_video_path = 'C:/Users/USER/Desktop/main-project/dataset/object-detection/output.mp4'


# Specify the start and end timestamps in seconds
start_timestamp = min(ts_updated_filtered)
end_timestamp = max(ts_updated_filtered)

# Crop the video from start_timestamp to end_timestamp
ffmpeg_extract_subclip(input_video_path, start_timestamp, end_timestamp, targetname=output_video_path)
```
